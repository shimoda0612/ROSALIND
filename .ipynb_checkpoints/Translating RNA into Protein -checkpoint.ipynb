```python
from Bio.Alphabet import IUPAC
from Bio.Seq import Seq
```

```python
def open_file():
    with open("rosalind_prot.txt",encoding = "utf-8") as f:
        data = f.read()
        #print(data)
    return data
```

```python
def translation(data) :
    seq = Seq(data, IUPAC.ambiguous_rna)
    print(seq.translate())
    
```

```python
if __name__ == "__main__" :
    data = open_file()
    translation(data)
```
